## Question 1: Use yfinance to Extract Stock Data
Reset the index, save, and display the first five rows of the tesla_data dataframe using the head function.

```python
import yfinance as yf
import pandas as pd
import json
import requests
from bs4 import BeautifulSoup
import yfinance as yf
import pandas as pd
import json
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
# Get data on this ticker
tesla_data = yf.Ticker('TSLA')

# Get the historical prices for this ticker
tesla_df = tesla_data.history(period='1d', start='2010-01-01', end='2023-03-18')

# Reset the index
tesla_df = tesla_df.reset_index()

# Save the data
tesla_df.to_csv('tesla_data.csv', index=False)

# Display the first five rows of the dataframe
print(tesla_df.head())
```

## Question 2: Use Webscraping to Extract Tesla Revenue Data
Display the last five rows of the tesla_revenue dataframe using the tail function. 

```python
# Define the URL for Tesla's revenue data
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the revenue data
table = soup.find('table', {'class': 'historical_data_table'})

# Convert the table to a dataframe
tesla_revenue = pd.read_html(str(table))[0]


# Clean up the dataframe
tesla_revenue.columns = ['Year', 'Revenue (Millions of USD)']
tesla_revenue['Revenue (Millions of USD)'] = tesla_revenue['Revenue (Millions of USD)'].str.replace(',', '', regex=True).str.replace('$', '', regex=True).astype(int)


# Display the last five rows of the dataframe
print(tesla_revenue.tail())
```

## Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

```python
# Define function to create a graph from a dataframe
def make_graph(df, title):
    plt.figure(figsize=(12, 6))
    plt.plot(df['Date'], df['Close'])
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Create a graph of Tesla's stock data
make_graph(tesla_df, 'Tesla Stock Data')
```
